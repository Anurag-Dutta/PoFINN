```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('text', usetex=True)
plt.rc('font', family='serif')

data_path = 'C:\\Users\\Anurag Dutta\\Desktop\\TIME SERIES\\c8.csv'
data = pd.read_csv(data_path)
timeT = data['Ageing Time'].values
degCAP = data['C8'].values

resampT = np.linspace(timeT[0], timeT[-1], len(timeT) * 100)
resampdegD = np.interp(resampT, timeT, degCAP)

output_path = data_path.replace('.csv', '_interpolated.csv')
interpolated_data = pd.DataFrame({'Ageing Time': resampT, 'C8': resampdegD})
interpolated_data.to_csv(output_path, index=False)

print(f"Interpolated dataset saved to {output_path}")

plt.figure()
plt.plot(timeT, degCAP, 'k*-', label='Original')
plt.plot(resampT, resampdegD, 'b--', label='Interpolated')
plt.xlabel(r'Ageing Time (hours)')
plt.ylabel(r'Capacitance (F)')
plt.legend()
plt.grid(True)

pdf_output_path = data_path.replace('.csv', '_plot.pdf')
plt.savefig(pdf_output_path, format='pdf', bbox_inches='tight')
print(f"Plot saved as PDF to {pdf_output_path}")

plt.show()
```
